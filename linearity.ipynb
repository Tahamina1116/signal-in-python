```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=np.arange(0,11)
x1=np.sin(.1*np.pi*n)
x2=np.cos(0.2*np.pi*n)

b=[2,3,1]
a=[1]
y1=np.convolve(b,x1,mode='same')
y2=np.convolve(b,x2,mode='same')

alpha=1.5
beta=0.5
x_sum=alpha*x1+beta*x2
y_sum=np.convolve(b,x_sum,mode='same')

is_linear=np.array_equal(y_sum,alpha*y1+beta*y2)

plt.subplot(3,1,1)
plt.stem(n,x1,basefmt="b")
plt.stem(n,x2,basefmt="r")
plt.title('input signal')
plt.xlabel('n')
plt.ylabel('amplitude')
plt.legend(['x1[n]','x2[n]'])

           

plt.subplot(3,1,2)
plt.stem(n,y1,basefmt="b")
plt.stem(n,y2,basefmt="r")
plt.title('output signal')
plt.xlabel('n')
plt.ylabel('amplitude')
plt.legend(['y1[n]','y2[n]'])
           

plt.subplot(3,1,3)
plt.stem(n,alpha*y1+beta*y2,basefmt="g")
plt.stem(n,y_sum,basefmt="k")
plt.title('linear combination')
plt.xlabel('n')
plt.ylabel('amplitude')
plt.legend(['alpha*y1[n]+beta*y2[n]','T{alpha*x1[n]+beta*x2[n]}'])
if is_linear:
    msg='the system is linear.'
else:
    msg='the system is not linear.'

plt.figtext(0.15,0.08,msg,bbox=dict(facecolor='white',alpha=1))
plt.tight_layout()
plt.subtitle('linearity verification',y=1.02)
plt.show()
           

plt.subplot(3,1,1)
plt.stem(n,y1,basefmt="b")
plt.stem(n,y2,basefmt="r")
plt.title('output signal')
plt.xlabel('n')
plt.ylabel('amplitude')
plt.legend(['y1[n]','y2[n]'])
           

if is_linear:
    msg='the system is linear'
else:
    msg='the system is not linear'
    plt.figtext(0.15,0.08,msg,bbox=dict(facecolor='white',alpha=1))
    plt.tight_layout()
    plt.suptitle('linearity verification',y=1.02)
    plt.show()
```
