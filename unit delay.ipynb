```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#given signal parameter
```

```python
n=np.arange(-10,11)
```

```python
#generate the signal x(n)
```

```python
x=np.zeros_like(n)
x[(n>=0)&(n<=4)]=4
x[(n>=-5)&(n<=-1)]=n[(n>=-5)&(n<=-1)]+4
```

```python
n_shifted_left=n+3
n_shifted_right=n-3
```

```python
x_shifted_left=np.zeros_like(n_shifted_left)
x_shifted_right=np.zeros_like(n_shifted_right)
x_shifted_left[(n_shifted_left>=0)&(n_shifted_left<=4)]=4
x_shifted_left[(n_shifted_left>=-5)&(n_shifted_left<=-1)]=n_shifted_left[(n_shifted_left>=-5) &(n_shifted_left<=-1)]+4
x_shifted_right[(n_shifted_right>=0)&(n_shifted_right<=4)]=4
x_shifted_right[(n_shifted_right>=-5)&(n_shifted_right<=-1)]=n_shifted_right[(n_shifted_right>=-5)&(n_shifted_right<=-1)]+4

                                                                       
```

```python
plt.figure()
```

```python
plt.subplot(3,1,1)
plt.stem(n,x,'r',basefmt='r',linefmt='r')
plt.title('original output signal x[n]')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.grid()
```

```python
plt.subplot(3,1,2)
plt.stem(n_shifted_left,x_shifted_left,'b',basefmt='b',linefmt='b') 
plt.title('original output signal x[n-3]')
plt.xlabel('n')
plt.ylabel('x[n-3]')
plt.grid()
```

```python
plt.subplot(3,1,3)
plt.stem(n_shifted_right,x_shifted_right,'b',basefmt='b',linefmt='b') 
plt.title('original output signal x[n+3]')
plt.xlabel('n')
plt.ylabel('x[n+3]')
plt.grid()
```
