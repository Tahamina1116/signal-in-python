```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T=2
D=0.7
N=10
```

```python
t=np.linspace(0,T,1000)
f_fourier=np.zeros_like(t)
```

```python
for n in range(1,N+1):
    k=2*n-1
    harmonic=(1/k)*np.sin(2*np.pi*k*t/T)
    f_fourier=(4/np.pi)*f_fourier
```

```python
plt.figure()
plt.plot(t,np.sin(np.sin(2*np.pi*t/T)),'b',linewidth=2)
plt.plot(t,f_fourier,'r--',linewidth=1)
plt.grid(True)
plt.title('fourier series for square wave')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.legend(['square wave','furier series'])
plt.show()
```
