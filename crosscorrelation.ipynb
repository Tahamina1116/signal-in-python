```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([0,0,2,-1,3,7,1,2,-3,0,0])
y=np.array([0,0,1,-1,2,-2,4,1,-2,5,0,0])
```

```python
cross_corr=np.correlate(x,y,mode='full')
```

```python
auto_corr=np.correlate(x,x,mode='full')
```

```python
property_verified=cross_corr[len(x)-1]=auto_corr[len(x)-1]
```

```python
plt.figure()
```

```python
plt.subplot(2,1,1)
plt.stem(cross_corr)
plt.title('cross-correlation of x[n] and y[n]')
plt.xlabel('Lag(k)')
plt.ylabel('amplitude')
```

```python
plt.subplot(2,1,1)
plt.stem(auto_corr)
plt.title('auto-correlation of  x[n]')
plt.xlabel('Lag(k)')
plt.ylabel('amplitude')
```

```python
if property_verified:
    print('property verified:The value at the specified lag matches in cross-correlation and autocorrelation')
else:
    print('property not verified:The value at the specified lag does not match in crosss-correlation and auto-correlation')
```
