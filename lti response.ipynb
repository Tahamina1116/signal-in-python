```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,6])
h=np.array([1,2,1,-1])
```

```python
y=np.convolve(x,h,mode='full')
n_x=np.arange(len(x))
n_h=np.arange(len(h))
n_y=np.arange(len(y))
```

```python
print("input signal x[n]:")
print(np.vstack((n_x,x)).T)
print("\nImpulse response h[n]:")
print(np.vstack((n_h,h)).T)
print("\nOutput Response y[n]:")
print(np.vstack((n_y,y)).T)
```

```python
plt.subplot(3,1,2)
plt.stem(n_h,h,'r')
plt.title('impulse signal h[n]')
plt.xlabel('n')
plt.ylabel('amplitude')
```

```python
plt.subplot(3,1,3)
plt.stem(n_y,y,'g')
plt.title('impulse signal y[n]')
plt.xlabel('n')
plt.ylabel('amplitude')
```

```python
plt.subplot(3,1,1)
plt.stem(n_x,x,'b')
plt.title('Input signal x[n]')
plt.xlabel('n')
plt.ylabel('amplitude')
```
