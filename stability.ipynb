```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
b=[1,-1,5,0.7]
a=[1]
```

```python
n=np.arange(0,51)
x=np.sin(0.1*np.pi*n)
y=np.convolve(b,x,mode='full')
p=np.roots(a)
```

```python
plt.subplot(2,1,1)
plt.plot(np.real(p),np.imag(p),'x')
plt.grid(True)
plt.title('Pole-Zero Plot')
plt.xlabel('real part')
plt.ylabel('imaginary part')
```

```python
if all(np.abs(p)<1):
    msg='the system is stable.'
else:
    msg='the system is not stable.'
    plt.figtext(0.15,0.45,msg,bbox=dict(facecolor='white',alpha=1)
```

```python
plt.subplot(2,1,2)
plt.stem(n,x,basefmt='b')
plt.stem(n,y[:len(n)],basefmt='r')
plt.title('input and output signal')
plt.xlabel('n')
plt.ylabel('amplitude')
plt.legend=(['input(x[n]','output(y[n])'])
```
