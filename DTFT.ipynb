```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Fs=1000
T=1/Fs
L=1000
t=np.arange(0,L)*T
f=50
```

```python
x=np.sin(2*np.pi*f*t)
```

```python
N=len(x)
X=np.fft.fft(x)/N
frequencies=np.fft.fftfreq(N,d=T)
```

```python
plt.figure()
plt.plot(frequencies,np.abs(X))
plt.title('Masgnitude of DTFT')
plt.xlabel('Frequency(Hz)')
plt.ylabel('|X(f)|')
plt.grid()
plt.show()
```
