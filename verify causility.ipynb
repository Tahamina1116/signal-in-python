```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=np.arange(0,11)
x=np.sin(.1*np.pi*n)
```

```python
b=[1,2,1]
a=[1]
```

```python
y=np.convolve(b,x,mode='full')
y_causal=y[:len(x)]
```

```python
plt.figure(figsize=(10,6))
```

```python
plt.subplot(2,1,1)
plt.stem(n,x,basefmt="b")
plt.title('input signal')
plt.xlabel('n')
plt.ylabel('x[n]')


plt.subplot(2,1,2)
plt.stem(n,y[:len(n)],basefmt="r")
plt.title('output signal')
plt.xlabel('n')
plt.ylabel('y[n]')
if np.all(y_causal[2:]==(x[2:]+2*x[1:-1]+x[:-2])):
 causality_msg='the system is causal.'
else:
 causality_msg='the system is not causal.'
plt.figtext(0.15,     0.85,  causality_msg,  bbox=dict(facecolor='white',alpha=1))
plt.legend(['output(y[n]'])
plt.tight_layout()
plt.subtitle('causality verification',y=1.02)
plt.show()
```
